# Modeling:

## 1. ARIMA Model:

```python
ndf.shape
```

```python
# Split the data into train and test
train = ndf[:300]
test = ndf[300:]
```

```python
from statsmodels.tsa.arima_model import ARIMA

arima_model = ARIMA(train, order=(1,1,1))  
model_fitting = arima_model.fit()  

print(model_fitting.summary())
```

```python
# Forecast
forcast_ar = model_fitting.forecast(steps=30, alpha=0.05)
```

```python
# Make as pandas series
fc_series = pd.Series(forcast_ar)
```

```python
# Plot
plt.figure(figsize=(20,10), dpi=70)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
ndf['forcast'] = model_fitting.predict()
ndf[['No_of_shows_added', "forcast"]].plot(figsize=(10,5))
```
